```python
import pyupbit
import numpy as np
```

```python
def get_ror(k):
    df = pyupbit.get_ohlcv("KRW-BTC", count=7)
    df['range'] = (df['high'] - df['low']) * k
    print('***df range:', df['range'])
    df['target'] = df['open'] + df['range'].shift(1)
    print('***df target:', df['target'])

    fee = 0.0005
    df['ror'] = np.where(df['high'] > df['target'],
                         df['close'] / df['target'] - fee,
                         1)

    ror = df['ror'].cumprod()[-2]
    return ror
```

```python
k_list = []
ror_list = []
for k in np.arange(0.01, 1.0, 0.01):
    ror = get_ror(k)
    print('*'*40)
    print("%.2f %f" % (k, ror))
    print('*'*40)
    k_list.append(k)
    ror_list.append(ror)
```
